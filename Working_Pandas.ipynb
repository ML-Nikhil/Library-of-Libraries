```python
# Creating new data sets

dict1 = {'name':['name1','name2','name3','name4','name5','name6'],'age':[12,20,24,20,18,14],}
dict2 = {'name':['name2','name3','name5','name7'],'edu':['X',"XII","X","IX"],}
```

```python
df1 = pd.DataFrame(dict1)
df2 =pd.DataFrame(dict2)
```

```python
# Joins - Left,Right,Inner, Outer

pd.merge(df1,df2,how='left',on='name')
```

```python
pd.merge(df1,df2,how='right',on='name')
```

```python
pd.merge(df1,df2,how='inner',on='name')
```

```python
pd.merge(df1,df2,how='outer',on='name')
```
